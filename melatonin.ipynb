We can estimate the sentiment based on rating and review_header to impute the missinng values.

Ratings:

1 - Very Poor

2 - Poor

3 - Neutral

4 - Good

5 - Very Good


```python
rating_dict = {1:"Very Poor Product",2:"Poor Product",3:"doesn't work",4:"Good Product",5:"Excellent Product"}
rating_dict
```

```python
# Imputing the missing values for the review_text column

def impute_review_text(row):
    if pd.isnull(row['review_text']) or pd.isna(row['review_text']):
        return rating_dict.get(row["review_rating"], "Unknown rating")
    else:
        return row['review_text']
```

```python
df['review_text'] = df.apply(impute_review_text, axis=1)
```

```python
# Imputing the missing values for the review_header column
def impute_review_header(row):
    if pd.isnull(row['review_header']) or pd.isna(row['review_header']):
        return rating_dict.get(row["review_rating"], "Unknown rating")
    else:
        return row['review_header']
```

```python
df['review_header'] = df.apply(impute_review_header, axis=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk

nltk.download('stopwords')
nltk.download('punkt')
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

column_transformer = ColumnTransformer(
    transformers=[
        ('one_hot_encoder', OneHotEncoder(), ["reviewed_variant_asin"])
    ],
    remainder='passthrough'
)


df["reviewed_variant_asin"].fillna("N/A", inplace = True)
r = column_transformer.fit_transform(df)
```

```python
# Removing the stopwords

def tokenize(text):
    stop_words = set(stopwords.words('english'))
    word_tokens = word_tokenize(text)
    filtered_text = [word for word in word_tokens if word.lower() not in stop_words]
    return ' '.join(filtered_text)

df['review_text'].fillna('N/A', inplace=True)
df['review_header'].fillna('N/A', inplace=True)
df['cleaned_review_text'] = df['review_text'].apply(tokenize)
df['cleaned_review_header'] = df['review_header'].apply(tokenize)

X = df[['cleaned_review_text', 'cleaned_review_header', 'average_rating', 'review_rating', 'sentiment_score']]
X = pd.concat([X.reset_index(drop=True), pd.DataFrame(r)], axis=1)
y = df['dose']

```

### Spliting the dataset into Train and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state=42)


vectorizer = TfidfVectorizer(max_features=250, max_df=0.8, stop_words='english')
X_train_tfidf_text = vectorizer.fit_transform(X_train['cleaned_review_text'])
X_train_tfidf_header = vectorizer.fit_transform(X_train['cleaned_review_header'])
X_test_tfidf_text = vectorizer.transform(X_test['cleaned_review_text'])
X_test_tfidf_header = vectorizer.transform(X_test['cleaned_review_header'])
```

```python
X_train_features = pd.concat([X_train.iloc[:, 2:17].reset_index(drop=True), pd.DataFrame(X_train_tfidf_text.toarray()), pd.DataFrame(X_train_tfidf_header.toarray())], axis=1)
X_test_features = pd.concat([X_test.iloc[:, 2:17].reset_index(drop=True), pd.DataFrame(X_test_tfidf_text.toarray()), pd.DataFrame(X_test_tfidf_header.toarray())], axis=1)

X_train_features.columns = X_train_features.columns.astype(str)
X_test_features.columns = X_test_features.columns.astype(str)

y_train = pd.to_numeric(y_train)
y_test = pd.to_numeric(y_test)
```

### Standardization of Numerical columns

### Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
model = LogisticRegression()
model.fit(X_train_features, y_train)

y_pred_logreg = model.predict(X_test_features)

print("Test Set")
print("accuracy score:", accuracy_score(y_test, y_pred_logreg))
print("classification report:")
print(classification_report(y_test, y_pred_logreg))
cm = confusion_matrix(y_test, y_pred_logreg,normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot()
plt.show()
print()
```

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(max_depth=30)
model.fit(X_train_features, y_train)

y_pred_rf = model.predict(X_test_features)

print("Test Set")
print("accuracy score:", accuracy_score(y_test, y_pred_rf))
print("Decision Tree report:")
print(classification_report(y_test, y_pred_rf))
cm = confusion_matrix(y_test, y_pred_rf,normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot()
plt.show()
print()
```

```python
model.max_depth
```

```python
def depth_model(max_depth):
  model = DecisionTreeClassifier(max_depth=max_depth)
  model.fit(X_train_features, y_train)
  y_pred_train = model.predict(X_train_features)
  train_acc = accuracy_score(y_train,y_pred_train)
  y_pred_rf = model.predict(X_test_features)
  test_acc = accuracy_score(y_test,y_pred_rf)
  print("Max Depth:", max_depth)
  print("Train Accuracy: ",train_acc)
  print("Test Accuracy:", test_acc)
  print()
  return(train_acc,test_acc)
```

```python
%timeit
train_error = []
test_error= []
max_depth = np.arange(1,31)

for i in range(1,31):
    train_acc, test_acc = depth_model(i)
    train_error.append(1-train_acc)
    test_error.append(1-test_acc)

```

```python
error_df = pd.DataFrame({'max_depth':max_depth,'train_error':np.array(train_error),'test_error':np.array(test_error)})
```

```python
error_df
```

```python
plt.figure(figsize=(10, 6))
plt.plot(error_df['max_depth'], error_df['train_error'], marker='o', label='Train Error')
plt.plot(error_df['max_depth'], error_df['test_error'], marker='x', label='Test Error')
plt.xlabel('Max Depth')
plt.ylabel('Error')
plt.title('Train and Test Errors vs. Max Depth')
plt.legend()
plt.grid(True)
plt.show()
```

We find after max depth of 10, the train and test accuracy remains almost same so we tune using max_depth 10

```python
model = DecisionTreeClassifier(max_depth=30)
model.fit(X_train_features, y_train)

y_pred_rf = model.predict(X_test_features)

print("Test Set")
print("accuracy score:", accuracy_score(y_test, y_pred_rf))
print("Decision Tree report:")
print(classification_report(y_test, y_pred_rf))
cm = confusion_matrix(y_test, y_pred_rf,normalize='true')
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=model.classes_)
disp.plot()
plt.show()
print()
```
